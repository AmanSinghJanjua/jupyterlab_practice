```python
def compute_mean(data_list):
    """ Calculates the arithmetic mean (average) of a list of numbers and prints the result. Handles empty lists by providing a message. """
    if not data_list:
        print("List is empty. Cannot compute mean.")
        return

    # Sums the values and divides by the number of elements (the list's length)
    mean_value = sum(data_list) / len(data_list)
    print(mean_value)

# Test case: The mean of 2, 4, 6, and 8 is (2+4+6+8)/4 = 20/4 = 5.0
compute_mean([2, 4, 6, 8])
```

This function takes a list of numbers, calculates the mean (average), and prints the result. It first checks if the list is empty to avoid division by zero. Then, it uses Python’s built-in `sum()` and `len()` functions to compute the mean.

Here are two ways to express the mean of a set of values in LaTeX:

1. Using summation notation:
   $$
   \bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i
   $$

2. Using a larger parenthesis for clarity:
   $$
   \bar{x} = \left( \frac{1}{n} \right) \sum_{i=1}^{n} x_i
   $$
